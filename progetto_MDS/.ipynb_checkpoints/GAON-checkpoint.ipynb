## Introduzione
L'analisi fattoriale esplorativa (EFA) è un metodo statistico usato per scoprire la struttura latente di un insieme relativamente grande di variabili. L'EFA è una tecnica all'interno dell'analisi dei fattori il cui obiettivo principale è quello di identificare le relazioni sottostanti tra le variabili misurate.

Il set di dati utilizzato è il risultato di una ricerca di mercato condotta da GRS Research & Strategy a Dubai, tra i visitatori di Expo 2020. Obiettivo della ricerca è quello di indagare opinioni e soddisfazione di residenti e visitatori nei confronti dell'offerta gastronomica della città.

#### Analisi delle componenti principali (PCA)

Si può provare a ridurre la dimensionalità dei dati eseguendo la PCA, tecnica di analisi multivariata che ha come obiettivo una riduzione della dimensionalità. Vuole cioè ridurre un numero elevato di variabili che descrivono un insieme di dati a un numero minore di variabili latenti, limitando il più possibile la perdita di informazioni.

```python
# Loading pandas dataframe as anndata 
adata = qp.AnnData(df_new)

# # log(x+1) transformation for all data
qp.pp.log1p(adata)

# Standardization scaling per feature
qp.pp.scale(adata)
```

```python
# pca & plotting variance ratio for each pc
qp.tl.pca(adata, svd_solver='auto')
qp.pl.pca_variance_ratio(adata, n_pcs=len(adata.var_names))
```

L'analisi fattoriale permette alle variabili che sono correlate tra loro ma largamente indipendenti da altri sottoinsiemi di variabili di combinarsi come fattori. Si riduce così un gran numero di variabili osservate (caratteristiche/dimensioni) a un numero più piccolo di componenti/fattori. Spesso, queste analisi dei fattori producono una definizione operativa per un processo sottostante utilizzando i contributi (loadings) della variabile osservata in un fattore.

Per ottenere gli autovettori (cioè i nostri fattori o componenti) e gli autovalori (cioè la proporzione della varianza complessiva spiegata) necessari per calcolare il loading (o saturazione) di ogni componente, si utilizza la funzione "sklearn.decomposition.PCA".

```python
# export to dataframe
df_new_anndata = adata.to_df()
```

```python
from sklearn.decomposition import PCA

# Apply PCA by fitting the good data with the same number of dimensions as features
n=4
pca = PCA(n) 

# Transform log_samples using the PCA fit above
pca_values = pca.fit_transform(df_new_anndata)
```

Si calcolano i loadings della PCA per ogni componente usando la seguente formula:

$Loadings = Autovettori.\sqrt{Autovalori}$

```python
pca_loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
pca_loading_matrix = pd.DataFrame(pca_loadings, columns=['PC{}'.format(i) for i in range(1, n+1)], 
                              index=df_new_anndata.columns)
pca_loading_matrix['Highest_loading'] = pca_loading_matrix.idxmax(axis=1)
pca_loading_matrix = pca_loading_matrix.sort_values('Highest_loading')
pca_loading_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(25,4))

# plot the heatmap for correlation matrix
ax = sns.heatmap(pca_loading_matrix.drop('Highest_loading', axis=1).T, 
                vmin=-1, vmax=1, center=0,
                cmap=sns.diverging_palette(220, 20, n=200),
                square=True, annot=True, fmt='.2f')

ax.set_yticklabels(
    ax.get_yticklabels(),
    rotation=0);
```

Dalla heatmap qui sopra vediamo alcuni modelli di loadings che spiegano ogni componente, ma non sembrano troppo evidenti. Questo perché l'obiettivo della PCA è quello di estrarre la massima varianza da un set di dati con poche componenti ortogonali, al fine di fornire solo una sintesi empirica del set di dati. Procediamo con l'analisi dei fattori per vedere se i loadings possono fornire un modello sottostante più chiaro per ogni fattore. 

#### Analisi fattoriale (FA)
Diversamente dalla PCA, il cui obiettivo è l'estrazione delle nuove componenti basandosi sulla correlazione tra variabili,  l’analisi fattoriale ha l'obiettivo di individuare fattori latenti.

#### Test per la fattorizzabilità di R 
Una matrice che è 'fattorizzabile' dovrebbe includere correlazioni consistenti. La dimensione prevista dipende, in una certa misura, da N (maggiori dimensioni del campione tendono a produrre correlazioni più piccole), ma se nessuna correlazione supera 0,30, l'uso di FA è discutibile perché probabilmente non c'è nulla da fattorizzare. 

Il test di sfericità di Bartlett è un test d’ipotesi che ha come ipotesi nulla quella che la matrice di correlazione coincida con la matrice identità. Se così fosse, non avrebbe senso performare una PCA in quanto le variabili incluse nell'analisi non sono per nulla correlate linearmente tra loro.
Il valore da calcolare per decidere se rifiutare o meno l’ipotesi nulla è il p-value. Perché il modello sia considerabile valido, il p-value deve essere inferiore a 0,05.

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
print('Bartlett-sphericity Chi-square: {}'.format(calculate_bartlett_sphericity(df_new_anndata)[0]))
print('Bartlett-sphericity P-value: {}'.format(calculate_bartlett_sphericity(df_new_anndata)[1]))
```

In alternativa, il test di Kaiser-Meyer-Olkin (la misura di Kaiser dell'adeguatezza del campionamento) è un rapporto tra la somma delle correlazioni quadrate e la somma delle correlazioni quadrate più la somma della correlazione parziale quadrata. Questo indice può assumere valori compresi tra 0 e 1 e, affinché abbia senso effettuare l'analisi, deve avere un valore almeno superiore a 0,5. Più il valore si avvicina ad 1, più l’adeguatezza del campione migliora. Questo indice può essere calcolato complessivamente per tutte le variabili incluse nella PCA.

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all, kmo_model = calculate_kmo(df_new_anndata[df_new_anndata.columns]);
print('KMO score: {}'.format(kmo_model));
```

#### Stima del numero di fattori e filtraggio delle variabili con comunalità > 0,2
In questo caso si è optato per una semplice analisi dei fattori senza alcuna rotazione.

```python
from factor_analyzer import FactorAnalyzer

fa = FactorAnalyzer(rotation=None)
fa.fit(df_new_anndata)
ev, v = fa.get_eigenvalues()
```

```python
ev
```

```python
# create scree plot using matplotlib
plt.scatter(range(1, 16), ev[:15])
plt.plot(range(1, 16), ev[:15])
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue/Variance ratio')
plt.grid()
plt.show()
```

Si osserva la presenza di un punto di "gomito" sul grafico, in prossimità di F1. Le comunalità delle variabili sono ispezionate per vedere se le variabili sono ben definite dalla soluzione. Le comunalità indicano la percentuale di varianza in una variabile che si sovrappone alla varianza nei fattori. Idealmente andrebbero abbandonate le variabili con basse comunalità, (es.: <0,2).

```python
communalities = pd.DataFrame(fa.get_communalities(), index=list(df_new_anndata.columns))
features_comm = list(communalities[communalities[0] > 0.2].index)
print('Variabili con comunalità > 0.2: {}'.format(len(features_comm)))
    
```

#### Analisi fattoriale con rotazione Varimax (ortogonale) e metodo di estrazione dei fattori Maximum Likelihood

```python
fa = FactorAnalyzer(rotation='varimax', n_factors=4, method='ml')
fa.fit(df_new_anndata);

# check eigenvalues
ev, v = fa.get_eigenvalues()
```

```python
fa_loading_matrix = pd.DataFrame(fa.loadings_, columns=['FA{}'.format(i) for i in range(1, 4+1)], 
                              index=df_new_anndata.columns)
fa_loading_matrix['Highest_loading'] = fa_loading_matrix.idxmax(axis=1)
fa_loading_matrix = fa_loading_matrix.sort_values('Highest_loading')
fa_loading_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(25,4))

# plot the heatmap per matrice di correlazione variabili/fattori
ax = sns.heatmap(fa_loading_matrix.drop('Highest_loading', axis=1).T, 
                vmin=-1, vmax=1, center=0,
                cmap=sns.diverging_palette(220, 20, n=200),
                square=True, annot=True, fmt='.2f')

ax.set_yticklabels(
    ax.get_yticklabels(),
    rotation=0);
```

```python
import seaborn as sns

plt.figure(figsize=(25,25))

# heatmap per matrice di correlazione fra variabili
corr = df_new_anndata.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

ax = sns.heatmap(corr, 
                 vmin=-1, vmax=1, center=0,
                 cmap=sns.diverging_palette(220, 20, n=200),
                 mask=mask, square=True, 
                 annot=True, fmt='.2f')
```
